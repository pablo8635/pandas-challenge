
## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results

```python
purchase = purchase_data.groupby(['Gender'])
all_purchase = pd.DataFrame(purchase.size().reset_index(name= 'Purchase Count'))
all_purchase = all_purchase.set_index('Gender')
average_p = pd.DataFrame(purchase['Price'].mean())
total_p = pd.DataFrame(purchase['Price'].sum())
all_purchase['Average Purchase Price'] = average_p
all_purchase['Total Purchase Price']= total_p
all_purchase['Average Purchase Price'] = all_purchase['Average Purchase Price'].map("${:,.2f}".format)
all_purchase['Total Purchase Price'] = all_purchase['Total Purchase Price'].map("${:,.2f}".format)
all_purchase
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24,29,34,39,100]
group_names = ['<10', '10-14', '15-19', '20-24','25-29','30-34','35-39','40+']
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
g1 = purchase_data.groupby(['Age Group'])
df = pd.DataFrame(g1.size())
df.columns = ['Total Count']
df['Percentage Players'] = round(100*(df['Total Count']/780))
df['Percentage Players'] = df['Percentage Players'].map("{:,.2f}%".format)
df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_p = purchase_data.groupby(['Item ID', 'Item Name'])
pc = most_p['Item ID'].count()
tpv = most_p['Price'].sum()
ip = most_p['Price'].sum()/most_p['Item ID'].count()

mpi_df = pd.DataFrame({'Purchase Count': pc, 'Item Price': ip, 'Total Purchase Value': tpv})
summary = pd.DataFrame(mpi_df.nlargest(5,'Purchase Count'))
top5 = summary[["Purchase Count","Item Price","Total Purchase Value"]]
top5
```
